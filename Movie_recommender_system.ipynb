```python
"""
The ast (Abstract Syntax Tree) module in Python is commonly used to parse and analyze Python source code.
Specifically, the function ast.literal_eval() is used to evaluate a string containing a literal Python expression,
such as a list, dictionary, tuple, or other basic data types, and return the corresponding Python object."""
import ast
```

```python
# Converting Genres and Keywords from Strings to Lists
```

```python
def get_genres(text):
    """ 
    this functin will provide all the genres list available in the genres column
    """
    l = [] # empty list for storing the genres
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'].apply(get_genres)
```

```python
# assigning into the column
movies['genres'] = movies['genres'].apply(get_genres)
```

```python
def get_keywords(text):
    """ 
    this function will provied the keywords available in the  column
    """
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['keywords'].apply(get_keywords)
```

```python
# assigning in the column
movies['keywords'] = movies['keywords'].apply(get_keywords)
```

```python
# fetching the cast list of top three actors
```

```python
def fetch_cast(text):
    """ 
    it will fetch the director name
    """
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'].apply(fetch_cast)
```

```python
movies['cast'] = movies['cast'].apply(fetch_cast)
```

```python
movies.head()
```

```python
# fetching the director name from the crew column
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def collapse(L):
    """
    is will joint the names and other strings because
    if we will not do that than the model will think that the string are same
    Ex: John Wayne and John Hurt are different person name
    so if we will not do that then model will think John is different
    """
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# applying on all these columns
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
def stem(text):
    """" this function will take the strings input and apply the stemming mechanism
        Ex: love loving and loved will be love
    """
    stemming_var = []
    for i in text.split():
        stemming_var.append(stemmer.stem(i))
    return " ".join(stemming_var) # again converting into string
```

```python
""" 
The sentence-transformers library is used to generate dense vector representations (embeddings) of sentences, paragraphs, or even larger chunks of text
"""
from sentence_transformers import SentenceTransformer
```

```python
# Load a pre-trained SentenceTransformer model specifically fine-tuned for paraphrase detection
# 'paraphrase-MiniLM-L6-v2' is a smaller, faster model that has been trained on paraphrase data to generate embeddings
# that capture semantic similarity and can be used for tasks like finding paraphrases or sentence similarity.


model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
# Apply the SentenceTransformer model to each row in the 'tags' column of the DataFrame 'df'
# The 'tags' column contains text data, and the model.encode(x) function generates a dense vector embedding
# for each text value in the 'tags' column. These embeddings are stored in a new column 'bert_embeddings'.

df['bert_embeddings'] = df['tags'].apply(lambda x: model.encode(x))
```

```python
# Compute the pairwise cosine similarity between all the sentence embeddings in the 'bert_embeddings' column
# We convert the embeddings into a list using `.tolist()` to pass it into the cosine_similarity function.
# The result is a square matrix where each element (i, j) represents the cosine similarity between the i-th and j-th embeddings.


cosine_sim = cosine_similarity(df['bert_embeddings'].tolist(), df['bert_embeddings'].tolist())
```

```python
def recommend(movie, top_n=5):
    # Step 1: Find the index of the movie in the dataset based on the title
    index = df[df['title'] == movie].index[0]

    # Step 2: Get cosine similarity scores for the selected movie against all other movies
    similarity_scores = list(enumerate(cosine_sim[index]))

    # Step 3: Sort movies by similarity score in descending order (most similar first)
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Step 4: Extract the titles of the top N similar movies, excluding the input movie itself
    recommended_movies = []
    for i in similarity_scores[1:top_n+1]:  # Start from index 1 to skip the input movie
        recommended_movies.append(df.iloc[i[0]]['title'])

    # Return the list of recommended movies
    return recommended_movies

# Example usage: Recommend 5 movies similar to 'Avatar'
movie_to_recommend = 'Avatar'
recommended_movies = recommend(movie_to_recommend, top_n=5)

# Print out the recommended movies
print(f"Movies similar to '{movie_to_recommend}':")
for idx, title in enumerate(recommended_movies, 1):
    print(f"{idx}. {title}")
```

```python
# checking

# Step 3: Define the movie recommendation function
def recommend(movie, top_n=5):
    # Step 1: Find the index of the movie in the DataFrame using the movie title
    # This retrieves the index of the movie within the DataFrame based on the 'title' column.
    index = df[df['title'] == movie].index[0]

    # Step 2: Retrieve the cosine similarity scores for the selected movie against all other movies
    # We look up the similarity scores for the movie at the found index in the cosine similarity matrix.
    similarity_scores = list(enumerate(cosine_sim[index]))  # Each item is a tuple (movie_index, similarity_score)

    # Step 3: Sort the movies by their similarity score in descending order (most similar first)
    # The sorted list will contain tuples of (movie_index, similarity_score) sorted by similarity_score.
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Step 4: Extract the top N most similar movies (excluding the input movie itself)
    recommended_movies = []
    for i in similarity_scores[1:top_n+1]:  # Skip the movie itself by starting from index 1
        # For each similar movie, append its title to the list of recommended movies
        recommended_movies.append(df.iloc[i[0]]['title'])

    # Return the list of recommended movie titles
    return recommended_movies
```
